## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import iqr
from scipy.stats import linregress
import numpy as np
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html
df = pd.merge(study_results, mouse_metadata, how = 'outer', on = 'Mouse ID') 
# Display the data table for preview
df.head()
```

```python
# Checking the number of mice.
n_mice = df['Mouse ID'].nunique()
n_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html#pandas.DataFrame.drop_duplicates
df_clean = df.drop_duplicates(subset= ('Mouse ID', 'Timepoint'))
df_clean
```

```python
# Checking the number of mice in the clean DataFrame.
n_mice = df_clean['Mouse ID'].nunique()
n_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
drug_stat_df = df_clean.groupby('Drug Regimen')
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean = drug_stat_df['Tumor Volume (mm3)'].mean()
median = drug_stat_df['Tumor Volume (mm3)'].median()
variance = drug_stat_df ['Tumor Volume (mm3)'].var()
stdev = drug_stat_df['Tumor Volume (mm3)'].std()
sem = drug_stat_df['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary dataframe.
tumor_summary_df = pd.DataFrame({'Mean': mean, 'Median': median, 'Variance': variance,
                                    'Standard Deviation': stdev, 'SEM': sem})

tumor_summary_df.head()


```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
gtimept = pd.DataFrame(df_clean.groupby('Mouse ID')['Timepoint'].max())

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
mergedclean = pd.merge(gtimept, df_clean, on=['Mouse ID', 'Timepoint'], how='left')
mergedclean
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_regimen = ['Capomulin']

# Create empty list to fill with tumor vol data (for plotting)
t_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
#https://www.kite.com/python/answers/how-to-convert-a-column-in-a-pandas-dataframe-to-a-numpy-array-in-python
#https://stackoverflow.com/questions/53337391/how-do-you-calculate-interquartile-range-iqr-correctly-using-python
#https://stackoverflow.com/questions/23228244/how-do-you-find-the-iqr-in-numpy
drug_df = mergedclean.loc[mergedclean['Drug Regimen'] == 'Capomulin']
drug_df
t_arr = drug_df.loc[:,'Tumor Volume (mm3)']
numbers = t_arr.values
upperq, lowerq = np.percentile(numbers, [75 ,25])
quart = iqr(numbers, rng=(25,75))
print(f'IQR for {drug_regimen}: {quart}')
    # Determine outliers using upper and lower bounds
lower_bound = lowerq - (1.5*quart)
upper_bound = upperq + (1.5*quart)
print(f'Lower Bound for {drug_regimen}: {lower_bound}')
print(f'Upper Bound for {drug_regimen}: {upper_bound}') 
outliers_count = ([(drug_df.loc[:,'Tumor Volume (mm3)'] >= upper_bound) | 
                 (drug_df.loc[:,'Tumor Volume (mm3)'] <= lower_bound)]).count()
print(f'Number of {drug} outliers: {outliers_count}\n')
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
cap_drug = df_clean.loc[df_clean['Drug Regimen'] == 'Capomulin']
cap_drug
```

```python
ave_cap =  cap_drug.groupby('Mouse ID').mean()
ave_cap
```

```python
x_values = ave_cap['Weight (g)']
y_values = ave_cap['Tumor Volume (mm3)']

plt.scatter(x_values, y_values, marker="o", facecolors="blue",
           edgecolors ="black")

plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume (mm3)')

plt.show()
```

## Correlation and Regression

```python
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html
x_values = ave_cap['Weight (g)']
y_values = ave_cap['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values, marker="o", facecolors="blue",
           edgecolors ="black")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,38),fontsize=15,color="red")
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Observations:

# 1) There is a positive correlation between a mouse's weight in grams and the tumor volume (mm3).

# 2) Mouse b128, treated with the Capomulin drug regimen, 
#showed a significant decrease in tumor volume between 
#datapoint 6 (timepoint 30 ) to datapoint 5 (timepoint 35). 
```
